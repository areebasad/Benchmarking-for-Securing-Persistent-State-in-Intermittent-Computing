```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from os import path
import pandas as pd
```

```python
textSizes = (4392, 16432, 17200)
dataSizes = (132,132,136)
bssSizes  = (8332, 8276,8832)
totalSizes = (12856, 24840, 26168)

men_means, men_std = (20, 35, 30, 35, 27), (2, 3, 4, 1, 2)
women_means, women_std = (25, 32, 34, 20, 25), (3, 5, 2, 3, 3)

ind = np.arange(3)  # the x locations for the groups len(textSizes)
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/4, textSizes, width,
                label='.text')
rects2 = ax.bar(ind + width/4, dataSizes, width,
                label='.data')
rects3 = ax.bar(ind + width/2, bssSizes, width,
                label='.bss')
rects4 = ax.bar(ind + width/2, totalSizes, width,
                label='Total')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Size in Bytes')
ax.set_xlabel('Hardware Vs Software Approach Code Size Comparison')
ax.set_title('Code sizes comparision')
ax.set_xticks(ind)
ax.set_xticklabels(('AES-Module (Hardware-based)', 'AES WolfSSL (Software-based)', 'AES mbedTLS (Software-based)'))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': -0.1, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom', size=8)


autolabel(rects1, "left")
autolabel(rects2, "right")
autolabel(rects3, "right")
autolabel(rects4, "right")

fig.tight_layout()

plt.show()
```
